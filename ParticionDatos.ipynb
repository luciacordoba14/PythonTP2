## Partición de Datos

### Módulos necesarios

```python
from sklearn.model_selection import train_test_split
import polars as pl
import pyprojroot

ROOT = pyprojroot.here()
```

Cargamos los datos de la temporada 1, los cuales se utilizarán para el entrenamiento de los modelos. Luego, estos datos se dividieron en un conjunto de entrenamiento y un conjunto de validación.

```python
datos_temporada1 = pl.read_parquet(ROOT / "Datos" / "datos_temporada1.parquet")
```

```python
entrenamiento, validacion = train_test_split(
    datos_temporada1.to_pandas(),
    test_size=0.2,
    random_state=123,
    stratify=datos_temporada1["Swing"].to_pandas(),
)

datos_entrenamiento = pl.from_pandas(entrenamiento)
datos_validacion = pl.from_pandas(validacion)
```

Guardamos ambos conjuntos para poder utilizarlos en otras notebooks. 

```python
datos_entrenamiento.write_parquet(ROOT / "Datos" / "datos_entrenamiento.parquet")
datos_validacion.write_parquet(ROOT / "Datos" / "datos_validacion.parquet")
```
